```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('KAts.csv', low_memory = False, parse_dates = ['Date'])
```

```python
df.head()
```

```python
df['Date']=pd.to_datetime(df['Date'].apply(lambda x: x+'-2020'))
```

```python
df.head()
```

```python
col = pd.Series(df.columns[4:])

def remove_X(x):
    return pd.to_datetime(x[1:])

c = col.apply(lambda x: remove_X(x))
c.index = range(len(df.columns))[4:]
```

```python
c
```

```python
c[0]=df.columns[0]
c[1]=df.columns[1]
c[2] = df.columns[2]
c[3] =df.columns[3]
```

```python
c.sort_index(inplace = True)
```

```python
df.columns = c
```
